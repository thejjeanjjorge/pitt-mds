# Pandas DataFrame Concatenation with `pd.concat`

This notebook demonstrates vertical and horizontal concatenation of Pandas DataFrames using `pd.concat`, along with common mistakes and best practices.

## Setup

```python
import pandas as pd
```

## Sample DataFrames

```python
df_a = pd.DataFrame({
    "id": [1, 2, 3],
    "value": ["A", "B", "C"],
})

df_b = pd.DataFrame({
    "id": [4, 5, 6],
    "value": ["D", "E", "F"]
})

print(df_a)
print(df_b)
```

## Vertical Concatenation

By default, `pd.concat` stacks DataFrames **row-wise** (vertically).  
The original index values are preserved, so index labels may repeat.

If the DataFrames do not have identical columns, Pandas automatically aligns
the columns and fills missing values with `NaN`.

```python
pd.concat([df_a, df_b])  #Notice the index values repeating.
```

Use `ignore_index=True` to reset the index after concatenation.

```python
pd.concat([df_a, df_b], ignore_index=True)
```

In the below example, `df_c` contains an additional column (`extra`) that does not exist in `df_a`.

When we concatenate the two DataFrames vertically using `pd.concat`, Pandas aligns
columns by name. Since `df_a` does not have the `extra` column, its rows are filled
with `NaN` values for that column.

```python
df_c = pd.DataFrame({
    "id": [1, 2, 3],
    "value": ["A", "B", "C"],
    "extra": [10, 20, 30]
})

pd.concat([df_c, df_a], ignore_index=True)
```

## Horizontal Concatenation

Setting `axis=1` concatenates DataFrames **column-wise**.  
The DataFrames are aligned by their index values.

If one DataFrame has more rows than the other, Pandas fills the missing
column values with `NaN`, similar to how mismatched columns are handled
during vertical concatenation.


```python
df_a
```

```python
df_d = pd.DataFrame({
    "score": [90, 85, 88]
})

pd.concat([df_a, df_d], axis=1)
```

### Horizonatal Concatenation with row mismatch

```python
df_e = pd.DataFrame({
    "id": [8, 9, 10, 11, 12],  # Two extra rows
})

pd.concat([df_a, df_e], axis=1)
```

## Handling Duplicate Column Names
Columns can repeat when concatenating horizontally. Rename columns beforehand to avoid confusion.

```python
df_a
```

```python
df_f = pd.DataFrame({
    "id": [8, 9, 10],
    "value": ["D", "E", "F"]
})
pd.concat([df_a, df_f], axis=1)
```

Note the repeating column names above, which can be quite confusing. To avoid this, let's rename columns before concatenating.

```python
df_f_renamed = df_f.rename(columns={"id": "extra_id", "value": "grade"})
pd.concat([df_a, df_f_renamed], axis=1)
```

## Summary
- `pd.concat` supports vertical and horizontal concatenation
- Indexes may repeat unless `ignore_index=True` is used
- When doing vertical/horizontal concatenation, NaN values will be filled accordingly.
- Manage column names carefully to avoid duplication
