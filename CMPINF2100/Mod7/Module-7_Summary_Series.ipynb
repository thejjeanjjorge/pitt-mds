# Calculating Summary Statistics of DataFrame Columns

This notebook demonstrates an important part of Exploratory Data Analysis (EDA): summarizing your data to understand:

- Typical values (averages)

- Spread and variability

- Frequency of categories

- Potential data quality issues

In Pandas, we often compute summary statistics on individual columns. Recall that columns of data in Pandas are represented as 1D Series objects.

## Setup

```python
import pandas as pd
```

## Selecting a Column

When you select a single column from a DataFrame, Pandas returns a Series.

```python
df = pd.DataFrame({
    "age": [21, 22, 21, 23, 22, 21],
    "major": ["CS", "DS", "CS", "Math", "DS", "CS"]
})

age_series = df["age"]
print(age_series)
type(age_series)
```

## Numerical Summary Statistics

### Mean, Min, Max

These give you the average, smallest, and largest values respectively in the column.

```python
print(age_series.mean())
print(age_series.min())
print(age_series.max())
```

### Standard Deviation

Note: Pandas’ `.std()` returns the unbiased estimator of the standard deviation, so no need to set `ddof=` as in NumPy. This is often what we want in statistics, unlike NumPy’s default behavior.

```python
age_series.std()
```

## Counting Unique Values

Counting the number of unique values in the column is useful for categorical or string variables.


```python
df["major"].nunique() # Number of unique majors.
```

```python
df["major"].unique() # To get the actual unique values present in the column
```

### Number of Rows vs Number of Unique Values

- `.size` -- total number of rows (elements) in the column

- `.nunique()` -- number of distinct values

These differ when values repeat.

```python
df["major"].size
```

```python
df["major"].nunique()
```

```python
print(df["major"].size)
print(df["major"].nunique())
```

## Counting Occurrences of Each Value

This is one of the most commonly used methods in Pandas. It's so nice! It gives you how many rows in a column have specific values.


```python
df["major"].value_counts()
```
