# Essential Exploratory Data Analysis (EDA) Methods and Identifying Missing Values with Pandas

This notebook introduces key Pandas methods for identifying missing values and performing some essential exploratory data analysis (EDA) for basic information on a dataset.

## Setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Example DataFrame with Missing Values
We'll look at a Formula 1 dataset of races and drivers. Note the `NaN` values where data is missing.

```python
df  = pd.DataFrame({
    "race_1": [25, np.nan, 18, 1],
    "race_2": [25, 12, np.nan, np.nan],
    "race_3": [25, 12, 18, np.nan]
}, index =  ["Max Verstappen", "Lewis Hamilton", "Fernando Alonso", "Lando Norris"])

df
```

## Identifying Missing Values

`df.col.isna()` or `df.col.isnull()` can be used to detect missing values. Both return a Boolean Series where:

- True indicates a missing value (`NaN`, `None`, or `NaT` for "not a time")

- False indicates a non-missing value

```python
df['race_3'].isna()
```

### Total Missing Values
You can quickly calculate the number of missing values in a column by running `.sum()` on `.isna()`.

```python
df['race_3'].isna().sum()
```

On certain versions of Pandas, you might see values returned like `np.int64(1)` instead of just `1`. That's fine: it's the data type as well as the value, which is whatever number is inside the parentheses.

You can also call `.isna().sum()` on an entire dataset, which will provide the number of missing values for every column.

```python
df.isna().sum() # returns total number of missing values in each column of the DataFrame.
```

### Proportion of Missing Values

```python
df.isna().mean() # returns mean number of missing values in each column of the DataFrame.
```

## Dropping Missing Values

- Drops rows that contain at least one missing value (NaN, None, NaT).

- Returns a new DataFrame (it doesn’t change the original dataframe unless you use inplace=True).

Doing this is sometimes referred to as keeping only the **complete cases**, i.e. rows where all columns have a value.



```python
complete_cases = df.dropna()
complete_cases
```

### Drop rows vs drop columns


axis=0 (default): drop rows

axis=1: drop columns

In the below example, each column has at least one missing value, so all columns are dropped!

```python
df.dropna(axis=1)
```

### Drop rows with missing values in particular columns
Sometimes you don't want to drop rows that have any missing value, but only rows with missing values in particular columns that may be important. In that case, you can use `.dropna(subset=[columns])`.

```python
df.dropna(subset=['race_2'])
```

The `how` parameter of `dropna` determines the strictness:

how="any" (default): drop row/column if any value is missing

how="all": drop row/column only if all values are missing

```python
df.dropna(how="all") # Doesn't delete anything since every row has one non-null value
```

## Essential EDA Methods
Here are some basic exploratory analyses that you'll want to know for almost all datasets you'll be working with. Before you get into further analysis, it's good to know:
* How many rows are in the dataset
* What columns are in the dataset
* What the data type of each column is
* The number of unique values per column
* The number of missing values per column

This information will often affect how you prepare the data for later visualization or analysis. For example, if there are lots of missing values in an important column, you might want to remove those missing values before analyzing the dataset. Or if there are just a handful of values for a continuous (numeric) column, you may want treat it as a categorical column instead.

### `.info()` for Quick DataFrame Information
The `df.info()` function provides much of this information at a glance, including the number of rows (entries), names of columns and their datatypes.

```python
df.info() # Returns information about a DataFrame including the index dtype and columns, non-NA values and memory usage.
```

`.info()` provides the number of non-null values for every column, which can lead to some mental gymnastics to see how many missing values there are per column, specifically subtracting the number of non-null values from the total number of rows. As discussed above, it's often nice to directly get the number of missing values for each column.

```python
df.isna().sum()
```

### Column Names and Data Types
It's often useful to get a list of the column names directly. `df.columns` is your go-to tool for that.

```python
df.columns # Returns the column labels of the DataFrame.
```

If you want that in list form, you can run:

```python
df.columns.tolist()
```

Other times you might want to get column data types directly.

```python
df.dtypes
```

### Number of Unique Values per Column

```python
df.nunique() #Returns the number of unique values per column.
```

By default, `NaN` is not included as a unique value. But you can include it by passing `dropna=False` to `nunique()`.

```python
df.nunique(dropna=False)
```

## Descriptive Statistics

Outside of the basic essential EDA, Pandas offers some handy functions for automatically getting information about the distribution of values within continuous (numeric).

`describe()` generates summary statistics for each continuous column in the DataFrame:
- count – number of non-missing values

- mean – average value

- std – standard deviation (measure of spread)

- min – minimum value

- 25% – first quartile (Q1) value

- 50% – median

- 75% – third quartile (Q3) value

- max – maximum value

```python
df.describe()
```

### Descriptive Statistics for Categorical Columns
Can you run `.describe()` for categorical columns? Yes, with the `include='object'` argument. It provides a different set of information more relevant to categorical columns:
* `count`: the count of non-null rows in the column
* `unique`: the number of unique values in the column
* `top`: the most frequent value
* `freq`: the most common value's frequency

Let's look at how this works with a dataframe with a categorical column.

```python
df_2 = pd.DataFrame({
    "race_1": [25, np.nan, 18, 1],
    "race_2": [25, 12, np.nan, np.nan],
    "race_3": [25, 12, 18, np.nan],
    "teams" : ["Red Bull", "Mercedes", "Ferrari", "McLaren"]
}
, index =  ["Max Verstappen", "Lewis Hamilton", "Fernando Alonso", "Lando Norris"])

df_2.describe(include='object')
```

Each value just had a count of 1 in the `teams` column, so Pandas chose the first one (Red Bull) as the top.

## Realistic Example: Titanic Dataset
Let's apply these techniques to a new dataset of Titanic passengers.

```python
titanic = sns.load_dataset('titanic')
titanic.info()
titanic.head()
```

We can see from `.info()` that there are 891 rows and 15 columns in the dataset. Column names and data types are also listed. From the first 5 rows and the Non-Null Count in `info()`, we can see that there are some missing (`NaN`) values. Let's see exactly how many.

```python
titanic.isna().sum()
```

Digging a bit deeper, let's see how many unique values we have for each column. We can treat continuous variables as categorical if they have a small number of unique values.

```python
titanic.nunique()
```

As a reminder, `.value_counts()` provides a useful way to obtain the number of rows per each unique value. Here we'll apply it to the `pclass` column, which is the class of the passenger's ticket (first, second, or third).

```python
titanic['pclass'].value_counts()
```

For continuous columns, we'll use .describe() to quickly obtain summary statistics. Note that these might not be as helpful for columns like `pclass` with only a few unique values, whether the passenger was in first, second, or third class. For those, it might be more helpful to know things like which is the most frequent value (as we investigated above).

```python
titanic.describe()
```

## Summary
- `isna()` and `isnull()` identify missing values. `isna().sum()` and `isna.mean()` quantify missing data
- `dropna()` returns complete cases
- Methods like `.info()`, `.nunique()`, and `.describe()` provide quick insights
