# Introduction to grouping, aggregation, and split-apply-combine

It's often useful to examine summary statistics for specific "groups" within a dataset. By groups, I mean observations that all share a value for a particular attribute. So if you have a dataset of tree measurements in a location, you might wonder the average height of maple trees, birch trees, and hickory trees. In this case, the tree species would be a group.

This process is sometimes called split-apply-combine, for the three steps involved:

1. **Split** the data into groups, that is, observations that share the same value for an attribute. 
2. **Apply** a function to each group. This could simply be a summary statistic like the mean or standard deviation for an attribute to summarize (different from the grouping attribute), or something more complicated.
3. **Combine** the resulting values for each group into an output dataset.

Here's a visual representation of this process from the [*Python Data Science Handbook*](https://jakevdp.github.io/PythonDataScienceHandbook/03.08-aggregation-and-grouping.html) by Jake VanderPlas. The function applied is `sum()`.



Thankfully Pandas provides a great function to automatically do all of that: `groupby()`! Let's take a look at how that works with a dataset of Titanic passengers.

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.info()
titanic
```

Let's first get summary statistics for the survival rate.

```python
titanic.survived.describe()
```

Now let's look at how the survival rate varies among classes of passengers with `groupby()`. You don't have to pass a particular column/variable to summarize over, but the syntax for `groupby()` with specifying a column to summarize over is:

`df.groupby(<grouping column>)[<summarizing column>].function()`

For example, to group by the passenger class and count the number of non-empty rows for each class:

```python
titanic.groupby('pclass')['survived'].count()
```

Now we can look at the average survival by class.

```python
titanic.groupby('pclass')['survived'].mean()
```

By default, `groupby()` drops `NaN` values. You can include `NaN` as a group by specifying the `dropna=False` argument.

```python
titanic.groupby('embark_town', dropna=False).survived.mean()
```
