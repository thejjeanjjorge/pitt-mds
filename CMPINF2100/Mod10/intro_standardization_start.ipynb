# Introduction to standardization

Soon we'll start predictive analytics in the course. For some of the techniques we'll be covering, it's useful to manipulate data before analyzing it, something called **preprocessing**. A key preprocessing technique is **standardization**, putting data across different continous variables on similar scales, or ranges.

Let's first visualize some raw data that has very different ranges across variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins') 
penguins.info() 
```

```python
sns.catplot(data=penguins, kind='box', height=3, aspect=2)
plt.show()
```

**Z-standardization** is a type of standardization that transforms data to a 0 mean and a standard deviation of 1. It does this by converting values of each feature to their z-score, the number of standard deviations from the mean. There are two steps to z-standardization:
1. Centering: subtract the mean from the value
2. Scaling: divide the centered value by the the sample standard deviation
