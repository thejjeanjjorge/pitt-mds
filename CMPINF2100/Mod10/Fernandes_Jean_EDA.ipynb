Your "Vibe" Predictors

To tell a story about the "vibe," you would look at how track_popularity changes when these values change:

danceability: How suitable a track is for dancing.

energy: A perceptual measure of intensity and activity.

valence: The musical positiveness (high valence = happy/cheerful, low valence = sad/angry).

acousticness: A confidence measure of whether the track is acoustic.
