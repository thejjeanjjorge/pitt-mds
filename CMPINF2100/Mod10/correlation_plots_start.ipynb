# Correlation plots

## Positive and negative correlation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load wine data
wine_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
wine_names = ['Cultivar', 'Alcohol', 'Malic_acid', 'Ash', 'Alcalinity_of_ash', 'Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanin', 'Color_intensity', 'Hue', 'OD280_OD315', 'Proline'] 
wine = pd.read_csv(wine_url, names=wine_names) 
wine.info() 
```

```python
sns.relplot(data=wine, x='Hue', y='Ash', height=3.5)
plt.show()
```

```python
sns.lmplot(data=wine, x='Hue', y='Ash', height=3.5)
plt.show()
```

The above data seems to be uncorrelated.

```python
sns.lmplot(data=wine, x='Alcalinity_of_ash', y='Total_phenols', height=3.5)
plt.show()
```

```python
sns.lmplot(data=wine, x='Total_phenols', y='Flavanoids', height=3.5)
plt.show()
```

# Correlation coefficients
Correlation coefficients provide a quantitative measure of the correlation between two variables. Correlation coefficients are between -1 and 1. A value close to -1 means there is a negative correlation between the variables, while a value close to 1 means there is a positive correlation.

Correlation coefficients among columns in Pandas dataframes are accessible with `dataframe.corr()`. This is sometimes referred to as a correlation matrix.

```python
wine_corr = wine[['Hue', 'Ash', 'Total_phenols']].corr()
wine_corr
```

# Correlation plots

Correlation plots use a heatmap to visualize correlations among variables.

```python
fig, ax = plt.subplots(figsize=(4,3))
sns.heatmap(wine_corr, vmin=-1, vmax=1, center=0, ax=ax)
plt.show()
```

**Diverging** color palettes are best for correlation plots since they show two different colors: one for values above the center (0) and another for values below the center. The intensity of the color shows represents how strong of a positive or negative correlation.

You can change the color palette (called the color map) with the `cmap` argument. More information and examples of color palettes available in Seaborn can be found in their [color palettes tutorial](https://seaborn.pydata.org/tutorial/color_palettes.html).

```python
fig, ax = plt.subplots(figsize=(4,3))
sns.heatmap(wine_corr, vmin=-1, vmax=1, center=0, ax=ax, cmap='coolwarm')
plt.show()
```

You can add the correlation coefficients to the plot with the `annot` argument.

```python
fig, ax = plt.subplots(figsize=(4,3))
sns.heatmap(wine_corr, vmin=-1, vmax=1, center=0, ax=ax, annot=True, cmap='coolwarm')
plt.show()
```
