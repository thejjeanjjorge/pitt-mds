# Styles and Color Palettes in Seaborn

With all the plotting we've been doing, you may have been wondering about controlling the visual aspects. This notebook will introduce you to a couple of the most powerful tools for adjusting color and style: specifying **color palettes** and **figure background styles**.

But that's just the tip of the iceberg! You can read more about figure aesthetics in the Seaborn [aesthetics documentation](https://seaborn.pydata.org/tutorial/aesthetics.html) and [in-plot object properties documentation](https://seaborn.pydata.org/tutorial/properties.html).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset

We will once again demonstrate on the Palmer Penguins dataset, which contains measurements for penguins.

```python
penguins = sns.load_dataset("penguins")
penguins.info()
penguins.head()
```

## Basic Scatterplot

First, let’s create a simple scatterplot between two continuous variables.

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm")
plt.show()
```

## Coloring by a Continuous Variable

We have already covered coloring points in a scatterplot by a categorical variable. You can also color the scatterplot using a **continuous variable**. Seaborn will shade the points based on intervals of the continuous variable, kind of like the bins in a histogram.

Here we'll color points by the continuous variable `body_mass_g`.

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g")
plt.show()
```

## Changing Color Palettes

A color palette is a pre-chosen set of colors that are used together to show variations. These can be specified by their name. Take a moment to scan Seaborn's [color palette documentation](https://seaborn.pydata.org/tutorial/color_palettes.html#) to see some of the options.

In particular, note the differences between the main types of color palettes:
* **qualitative** color palettes show separate categories with different colors, good for categorical variables
* **sequential** color palettes show shades within one or two colors, good for continuous variables
* **diverging** color palettes show shades within two colors, good for continuous variables that diverge from a central point (like 0)

Some palettes are designed to be accessible for people with color vision deficiencies.

Examples include:
- `colorblind`
- `viridis`

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g", palette="viridis")
plt.show()
```

The difference occurs because `body_mass_g` is a continuous variable, while the `colorblind` palette is designed for categorical data as a qualitative color palette.

For `body_mass_g`, try some other **sequential color palettes** such as:

- `mako`

These palettes create a smooth gradient, making it easier to see how values change across the scatterplot.

## Reversing Color Palettes

You can reverse a color palette by adding `_r` to the palette name.

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g", palette="mako_r")
plt.show()
```

Let's compare with the regular `mako` palette:

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g", palette="mako")
plt.show()
```

## Changing Figure Background Style

Seaborn allows you to change the background theme using `sns.set_style()`.

You can see available figure style themes [here](https://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-styles).

```python
sns.set_style("dark")

sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g")
plt.show()
```

## White grid theme

Sometimes it's nice to have grid lines to more precisely ascertain where points are positioned on the figure.

```python
sns.set_style("whitegrid")

sns.relplot(data=penguins,x="flipper_length_mm",y="bill_depth_mm",hue="body_mass_g")
plt.show()
```

##  Default Style

The default style is a **white background without grid lines**. This theme is simply the `white` theme.

```python
sns.set_style("white")
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="body_mass_g")
plt.show()
```
