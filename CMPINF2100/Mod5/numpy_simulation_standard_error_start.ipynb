# Simulate the standard error of the mean (SEM) with NumPy
It's a lot easier than with base Python

```python
import numpy as np
```

## Define the random number generator

```python
rg = np.random.default_rng(2100)
```

## Simulation
Let's generate random numbers organized into a 2D array. Specifically, we'll draw 7 random numbers from a uniform distribution between 0 and 1 and replicate that process 5000 times. We'll store each random sample as a row in the array of simulated numbers, which will have 5000 rows and 7 columns.

```python
n_samples = 7
n_replications = 5000
```

```python
random_arr = rg.random((n_replications, n_samples))
random_arr.shape
```

```python
random_arr[:5, :]
```

## Calculate the average of each sample of 7 random numbers

Calculate the mean of each row, that is, the means across columns. Columns are axis 1.

```python
# columns are exis 1
sample_mean = random_arr.mean(axis=1)
sample_mean.shape
```

## Calculate the standard error of the mean
The SEM is the standard deviation of the averages of the repeated samples

```python
sem = sample_mean.std(ddof=1)
sem
```

## Simulation and calculation of the SEM in 1 line of code

```python
rg.random((n_replications, n_samples)).mean(axis=1).std(ddof=1)
```

## Study the effect of the sample size on the standard error
Let's simulate taking 5000 repeated random samples across different sample sizes. Then we can see how the SEM varies by sample size.

```python
sample_sizes = 10 * np.arange(1, 11)
sample_sizes
```

```python
n_replications = 500
sem_vs_sample_size = [rg.random((n_replications, n_samples)).mean(axis=1).std(ddof=1) for n_samples in sample_sizes]
sem_vs_sample_size
```

Let's plot how the standard error varies by the sample size of our simulations.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(sample_sizes, sem_vs_sample_size, '-o')
ax.set_xlabel('sample size')
ax.set_ylabel('standard error of the mean')
plt.show()
```
