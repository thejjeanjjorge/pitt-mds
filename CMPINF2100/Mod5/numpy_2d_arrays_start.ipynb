# 2-dimensional (2D) NumPy arrays

```python
import numpy as np
```

## Creating 2D NumPy arrays from lists of lists

```python
list_of_lists = [
    [1, 'apple', 25],
    [2, 'banana', 36]
]
print(len(list_of_lists[0]))
```

```python
arr2d = np.array(list_of_lists)
```

```python
arr2d.ndim
```

```python
arr2d.shape
```

## Creating 2D arrays from 1D arrays

```python
all_zeros = np.zeros(5)
all_ones = np.ones(5)
```

```python
vstacked = np.vstack([all_zeros, all_ones])
vstacked
```

```python
vstacked.shape
```

We can "stack" the 1D arrays vertically on top of each other with `np.vstack()`

## Slicing and indexing 2D NumPy arrays

```python
my_2d = np.array(
    [[0,0,0],
     [1,1,1],
     [2,2,2],
     [3,3,3]]
)
my_2d
```

The syntax for slicing and indexing 2D NumPy arrays is `array[<row_index>, <column_index>]`  
You can use `:` to mean "all rows" or "all columns"

```python
my_2d[0, :] # row 0, all columns
```

```python
my_2d[-1, :]
```

```python
my_2d.size
```

```python
my_2d[1, 2]
```

```python
my_2d[:2, :] # first 2 rows, all columns
```

```python
my_2d[:3, 1:] # first 3 rows, columns starting at column 1
```
