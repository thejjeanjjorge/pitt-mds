# What plot type should you use?

It depends! There are different plots for different types of data and purposes. The type of plot is mostly determined by those "essential" EDA information you've practiced:

* data type of variables
* number of unique values for variables
* whether you want to plot **individual** variables or **relationships** between multiple variables

```python
import pandas as pd
import seaborn as sns
```

Let's load some data to illustrate some different plot types. Specifically, we'll load the GapMinder dataset with basic country information across the years, from the *Pandas for Everyone* book.

```python
gap_url = 'https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv'
gap = pd.read_csv(gap_url, sep='\t')
gap.info()
gap
```

## Influence of the variable data type

There are two main data types for variables for our purposes:
* Categorical: non-numeric values such as strings
* Continuous: numeric values such as integers and floats

```python
gap.dtypes
```

Note that continuous variables with few unique values can be treated as if they are categorical variables. If a continuous column has <20 values, it might make more sense to treat each value as a "category" instead of a continuous, numeric value. It's up to you as the analyst!

```python
gap.nunique()
```

## Basic plots for categorical and continuous variables

For categorical variables, start by plotting the number of rows for each value with a **bar chart**.

```python
gap.continent.value_counts().plot(kind='bar')
```

For continuous variables, start by plotting the distribution of values with a **histogram**.

```python
gap.lifeExp.hist()
```

## Plotting individual variables and relationships between variables

The plots we've done so far visualize distributions of values for just one variable. These are sometimes called **marginal** distributions and plots. We ignore all the other variables and just focus on one.

For choosing plot types, it's important to consider whether you want to plot just one variable or the relationships between two or more variables. **Relationship** plots examine how multiple variables change or vary together.
