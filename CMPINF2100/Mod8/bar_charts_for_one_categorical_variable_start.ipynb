# Bar charts for one categorical variable
Let's dive deeper into marginal plots (for one variable) for **categorical** variables. As a reminder, these are variables that are non-numeric and often have string data types.

The central question we'll address is, How many rows have each value for a particular categorical variable?

We'll introduce the Seaborn plotting package here for plotting these counts of categorical variable values.

## Seaborn
Seaborn is a plotting package that works well with Pandas dataframes. It uses Matplotlib to create plots "under the hood". There are two main kinds of functions in Seaborn:

1. Figure-level functions
    * Goal-oriented instead of about individual kinds of plots
    * Good for quick observations and getting started with EDA
    * Can easily get multiple facets (subplots)
    * Less control over the exact plots that are created
2. Axes-level functions
    * Create specific kinds of plots
    * Allow more control over the exact plot details

We'll mostly be using figure-level functions, but it's fine to use axes-level functions, too.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
penguins.info()
penguins
```

The figure-level function for plotting categorical variables is `sns.catplot()`. Let's use this to plot the counts of different values for categorical variables in this dataset of penguins.

```python
sns.catplot(data=penguins, x='species', kind='count', height=3)
```

The `height` argument specifies the height of each facet (subplot) in inches, while `aspect` specifies the width as `aspect` * `height`.

```python
sns.catplot(data=penguins, x='island', kind='count', height=3, aspect=2)
```

To just show the plot in Jupyter notebook and not the strange text describing the plot object in the output, you can use Matplotlib's `plt.show()`:

```python
sns.catplot(data=penguins, x='sex', kind='count', height=2)
plt.show()
```

The figure-level function `catplot(kind='count')` is calling the axes-level function `countplot()`, which you can call directly with Seaborn, too.

```python
sns.countplot(data=penguins, x='sex')
```

Axes-level functions like `countplot()` don't have convenient arguments for setting the figure size. But you can use the underlying Matplotlib functions for that.

Just change `x` to `y` to make a horizontal bar chart and plot the data on the y-axis in Seaborn.
