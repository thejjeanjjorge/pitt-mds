4. The CSV file named `'cereal.csv'` is read into the Pandas Dataframe `df` for you below. Using SEABORN, plot a **histogram** showing the distribution of ratings for all cereals. Do not use the bins argument. Instead, set `binwidth=10` and `binrange=(0, 100)`.

```python
sns.displot(data=df, x='rating', kind='hist', binwidth=10, binrange=(0, 100))
plt.title('Distribution of Ratings')
plt.show()
```

# 5-6 Combining Two Categorical

5. The CSV file named `'stars.csv'` is read into the Pandas Dataframe `df` for you below. Using SEABORN, plot a **grouped bar chart** showing the relationship between the `type` and `color` of the stars. You must use `x=type` and `hue=color` for the arguments. Display the plot. Then, change the default palette using `palette=color_palette` and plot width using `aspect=2`.

Which two colors are the most frequent for Supergiant type stars?

```python
# STARTER
color_palette = {'Red': 'red','Blue': 'blue','White': 'grey','Yellow': 'yellow','Orange': 'orange'}

df = pd.read_csv("stars.csv")
```

```python
sns.catplot(data=df, x="type", hue="color", kind="count")
plt.show()
```

```python
sns.catplot(data=df, x="type", hue="color", kind="count", palette=color_palette, aspect=2)
plt.show()
```

The most frequent colors for Supergiant are red and blue.

6. The CSV file named `'stars.csv'` is read into the Pandas Dataframe `df` for you below. Using SEABORN, plot a **heatmap** showing the relationship between the `type` and `spectral class` of the stars. You must use `pd.crosstab()` for the data argument. 

What is the most frequent spectral class for White Dwarf type stars?

```python
ct = pd.crosstab(df["type"], df["spectral class"])
ct
```

```python
sns.heatmap(ct, annot=True)
plt.show()
```
