# Using Pandas to Plot Bar Charts and Histograms

We've gone through plotting with Matplotlib and Seaborn, but Pandas also has some direct plotting functionality that uses Matplotlib by default. This can be useful to quickly create simple plots. This notebook will guide you through plotting bar charts for categorical variables, and histograms for continuous variables, using Pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Making Bar Plots in Pandas

To create a bar plot from Pandas, simply call `.plot.bar()` on a `Series` of data (a column of a `DataFrame` or the output of a function like `.value_counts()`. You can also run `.plot(kind='bar')` if you prefer.

We will illustrate this with a simple DataFrame containing categorical data.

```python
df  = pd.DataFrame({
    "race_1": [25, np.nan, 18, 1],
    "race_2": [25, 12, np.nan, np.nan],
    "race_3": [25, 12, 18, np.nan],
    "team" : ["Red Bull", "Mercedes", "Red Bull", "Mclaren"],
    "name": ["Max Verstappen", "Lewis Hamilton", "Fernando Alonso", "Lando Norris"]
})

df
```

```python
df["team"].value_counts().plot.bar(title="Team Count")
plt.show()
```

## Customizing Bar Plots from Pandas
There are many arguments to the Pandas `.plot()` method (see [documentation here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot)). We'll illustrate a few useful ones below.

```python
df["team"].value_counts().plot.bar(title="Team Frequency",
                              xlabel="Team", ylabel="Count", rot=45, 
                               figsize=(3.5, 3.5))
plt.show()
```

If you want even more control and flexibility (such as the ability to plot multiple datasets on the same plot) with Pandas, you can create Matplotlib `Figure` and `Axes` objects and plot the data from Pandas on them. Then you can modify the `Figure` and `Axes` objects as needed.

Here's the same plot above illustrating that method.

```python
fig, ax = plt.subplots(figsize=(3.5, 3.5))
df["team"].value_counts().plot.bar(ax=ax) # Plot on the ax object created above
ax.set_title("Team Frequency")
ax.set_xlabel("Team")
ax.set_ylabel("Count")
ax.tick_params(axis="x", labelrotation=45);
plt.show()
```

## Plotting Histograms with Pandas
You can quickly and simply plot histograms for looking at the distribution of a continuous variable in Pandas with `plot.hist()` or simply `.hist()`. Let's load some data to illustrate this.

```python
gapminder = pd.read_csv('https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv', sep='\t')
gapminder.info()
```

```python
gapminder.lifeExp.plot.hist()
plt.show()
```

You can also run `.hist()` on an entire DataFrame to plot histograms of all continuous variables in the DataFrame.

```python
gapminder.hist()
plt.show()
```
