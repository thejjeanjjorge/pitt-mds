# Introduction to `matplotlib`

Steps to plot data with `matplotlib`:

1. Define the figure and axes objects
2. Plot the data within the axes objects
3. Modify any figure or axes object attributes if needed
4. Show the figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Define figure and axes objects

```python
fig, ex = plt.subplot()
```

The `Figure` object is the entire figure that one or multiple plots can fit into.



*Figure from [Real Python](https://realpython.com/python-matplotlib-guide/#:~:text=A%20Figure%20object%20is%20the,%E2%80%9D%20as%20we%20might%20expect)*

You can easily create multiple subplots, confusingly called `Axes` data types in Matplotlib.

```python
fig, axes = plt.subplots(2)
```

```python
fig, axes = plt.subplots(2, 3) # 2 rows, 3 columns of subplots
```

## Step 2: Plot the data
On the `Axes` objects we created. We'll generally use other packages to do this like Seaborn, but it's still good to know as Seaborn is using Matplotlib to plot data.

```python
fig, axes = plt.subplots(2)
```

```python
sine_data = np.sin(np.linspace(0, 10, 100))
sine_data
```

```python
axes[0].plot(sine_data)
fig
```

```python
cos_data = np.cos(np.linspace(0, 10, 100))
cos_data
```

```python
axes[1].plot(cos_data)
fig
```

## Step 3: Modify any figure or axes object attributes if needed
To control the appearance, for example.

One option is controlling the figure size with the `figsize=(<horizontal size>, <vertical size>)`.

```python
fig, axes = plt.subplots(2, figsize=(4, 4))
axes[0].plot(sine_data)
axes[1].plot(cos_data)
fig
```
