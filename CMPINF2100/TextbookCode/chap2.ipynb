```python
x = 11
y = 7
z = 14

# 1. Create a list of the numbers
numbers = [x, y, z]

# 2. Use a list comprehension to keep only the odd numbers
# This uses the modulo operator (%) we discussed
odds = [n for n in numbers if n % 2 != 0]

# 3. Check if we found any odd numbers and print the result
if odds:
    print("The largest odd number is:", max(odds))
else:
    print("None of the numbers are odd.")
```

```python
x = 120
ans = 0
itersLeft = x
while (itersLeft != 0):
    ans = ans + x
    itersLeft = itersLeft - 1
print(str(x) + '*' + str(x) + '=' + str(ans))
```

```python
# Find a positive integer that is divisible by both 11 and 12
x = 1
while True:
    if x%11 == 0 and x%12 == 0:
        break
    x = x + 1
print(x, "is divisible by 11 and 12")
```
