# Scatterplots
For visualizing continuous-to-continuous variable relationships.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Penguins dataset
penguins = sns.load_dataset('penguins')
penguins.info()
```

The figure-level function for showing the relationship between continuous variables is `sns.relplot()`. The underlying axes-level function is `sns.scatterplot()`.

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', kind='scatter', height=3)
plt.show()
```

```python
sns.relplot(data=penguins, x='body_mass_g', y='bill_length_mm', kind='scatter', height=3)
plt.show()
```

## Incorporating categorical variable
You can use scatter plots to visualize 3 or more variables: 2 continuous ones and up to 3 categorical variables with color, row facets and column facets.

```python
penguins.info()
```

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species',kind='scatter', height=3)
plt.show()
```

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='island',kind='scatter', height=3)
plt.show()
```

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', col='species',kind='scatter', height=3)
plt.show()
```

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='island',kind='scatter', height=3)
plt.show()
```

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='island',kind='scatter', row='sex', height=3)
plt.show()
```
