# Point plots
If you find yourself wondering, "Do averages of a continuous variable change across categories?", then point plots are for you!

Point plots allow quick visualization of how means of a continuous variable vary across when conditioned, or grouped by, different values of a categorical variable. Let's take a look!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Gapminder dataset
gap_url = 'https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv' 
gap = pd.read_csv( gap_url, sep = '\t' ) 
gap.info() 
```

```python
sns.catplot(data=gap, x='year', y='lifeExp', kind='point', linestyle='none', height=3, aspect=2)
```

Lots of styling options can be found in the [`sns.pointplot` documentation](https://seaborn.pydata.org/generated/seaborn.pointplot.html). That's the axes-level function being run by `sns.catplot(kind='point')`.

```python
sns.catplot(data=gap, x='year', y='lifeExp', kind='point', linestyle='none', 
            capsize=0.1, err_kws={'linewidth' : 1}, height=3, aspect=2)
```

You can also compare 3 variables with point plots: 1 continuous and 2 categorical variables. One categorical variable can be separated on the x-axis, while the other can set the color.

```python
sns.catplot(data=gap, x='year', y='lifeExp', hue='continent', kind='point', linestyle='none',
            capsize=0.1, err_kws={'linewidth' : 1}, height=3, aspect=2)
plt.show()
```

## Error bars
So what are those error bars anyway? They represent the 95% confidence interval, which is calculated from the standard error of the mean! Specifically, the upper and lower bounds of the confidence intervals are 2 times the standard error of the mean from the estimated mean.

The range of the confidence interval can be adjusted.

```python
sns.catplot(data=gap, x='year', y='lifeExp', hue='continent', kind='point', linestyle='none',
            errorbar=('ci', 68), capsize=0.1, err_kws={'linewidth' : 1}, height=3, aspect=2)
plt.show()
```
