# Violin Plots: Visualizing Distribution Across Categories

A **violin plot** can be useful when we want to understand how the distribution of a continuous variable changes across categories.

<span style="font-size:55px;">🎻</span>

A violin plot combines multiple visual elements:

- A **Kernel Density Estimate (KDE)** on each side, showing the distribution shape
- A **mini boxplot** inside the violin, including the **median** and **interquartile range (IQR)** (the middle 50% of the data)

Violin plots are an alternative to boxplots that show the **shape of the distribution**.

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Example 1: Diamonds Dataset

We first use Seaborn’s `diamonds` dataset.

We'll use violin plots to compare the **distribution of diamond prices** across **diamond colors**.

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

## Basic Violin Plot

This plot shows the distribution of `price` across the different `color` categories.

```python
sns.catplot(data=diamonds,x="color",y="price",kind="violin",aspect=2)
plt.show()
```

Notice the kernel density estimation curve for each color, wider sections indicate higher data density, while narrower areas show lower density revealing skewness. The skewed nature of these distributions, with many more datapoints at lower prices, would not be captured by boxplots.

However, what's that little black box inside the violins? It's a tiny boxplot! You can still get some good information that boxplots provide from these internal boxplots within violin plots:

- White line: Represents the median.
- Whick center bar: Represents the interquartile range (IQR), containing 50% of the data.
- Thin lines (whiskers): Represent the rest of the distribution.

## Showing Quartiles Inside the Violin

We can add **horizontal lines for quartiles** using `inner="quartile"`.

This shows:
- The **median**
- The **25th percentile**
- The **75th percentile**

```python
sns.catplot(data=diamonds,x="color",y="price",kind="violin",aspect=2,inner="quartile")
plt.show()
```

## Why Use Violin Plots?

Compared to boxplots, violin plots show:

- The **shape of the distribution**
- Whether the data is **skewed**
- Whether there are **multiple peaks (multimodal distributions)**

## Example 2: GapMinder Dataset

Now we load a GapMinder dataset and visualize how **life expectancy changes across years**.

```python
gap_url = "https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv"
gap_df = pd.read_csv(gap_url, sep="\t")
gap_df.info()
gap_df.head()
```

## Violin Plot of Life Expectancy Across Years

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="violin", aspect=2)
plt.show()
```

## Interpretation

This plot allows us to see:

- How **life expectancy distributions change over time**
- How the **distribution (spread) of life expectancy changes across time**

In this case, we can see how not only has the median life expectancy increased over time, but the shape of the distributions has shifted to having a dominant peak at the higher end of the distribution.
