# Continuous-to-Continuous Relationships: Scatterplots with Regression Lines

In this notebook, we will add **trend lines** to scatterplots in order to visualize the relationship between two continuous variables.

Trend lines are also called:
- **best-fit lines**
- **regression lines**
- sometimes informally, **trend plots**

We'll talk much more about these regression lines later in the course when we get to predictive modeling. For now, just think of them as a line that presents the overall trend in the relationship between two continuous variables. As the variable on the x-axis increases, does the variable on the y-axis increase, decrease or stay about the same? This is what a trend line shows.

These lines help us see whether there is an approximately **linear relationship** between two variables.

We will use Seaborn’s **figure-level** plotting function to plot both a scatterplot of the data, and a regression line with it.

```python
sns.lmplot(...)

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Load the Data

We’ll use the Seaborn `penguins` dataset.

````python
penguins = sns.load_dataset("penguins")
penguins.head()
````

````python
penguins.info()
````

## A Simple Trend Line with `sns.lmplot()`

We start by plotting the relationship between:

- `flipper_length_mm`
- `bill_depth_mm`

Seaborn automatically:
- draws the scatterplot
- fits a linear trend line

````python
sns.lmplot(data=penguins,x="flipper_length_mm",y="bill_depth_mm")
plt.show()
````

Looking at the overall dataset, the trend line suggests a **negative relationship** between bill depth and flipper length: as **flipper length increases**, **bill depth appears to decrease**.

## Grouping by a Categorical Variable

However, sometimes the relationship between two continuous variables changes when we **condition on** or **group by** a categorical variable. This is comparing the relationship among 3 variables: 2 continuous variables and 1 categorical variable. For example, with the penguins dataset, we could see how the relationship between `flipper_length_mm` and `bill_depth_mm` may vary depending on the `species` of penguin.

You can use the `hue=` argument of `sns.lmplot` to color the scatterplot points and corresponding regression lines by unique values of a categorical variable.

````python
sns.lmplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="species")
plt.show()
````

Look at that! This is the same data we plotted before, but now our interpretation changes significantly.

When we group by species, the relationship between `flipper_length_mm` and `bill_depth_mm` now appears **positive within each species**. Within each species, as flipper length increases, bill depth also tends to increase.

However, when we plotted the overall relationship without grouping by species, we saw an **overall negative relationship** between `flipper_length_mm` and `bill_depth_mm`. That overall negative relationship in the first plot hides the positive within-group relationships.

This is value information for data analysis more generally: grouping by categories can change how we interpret a relationship between continuous variables.
