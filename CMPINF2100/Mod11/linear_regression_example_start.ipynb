# Linear regression example
Let's work through an example to illustrate what's going on with linear regression.

We'll use the example of predicting river depth from rainfall. Our outcome variable $y$ is river depth (in millimeters), where our input variable $x$ is rainfall (also in millimeters).

Here's the formula for predicting the output (actually the estimated mean output) from the input:

$$ \hat y = \beta_0 + \beta_1x $$

There are two coefficients in this predictive model: $\beta_0$ (the intercept) and $\beta_1$ (the slope). We'll assume they have already been estimated from data, something we'll do in the next module:
$$\beta_0 = 4780 $$
$$\beta_1 = 2.89 $$

Let's visualize this regression model as a line!

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

First we'll build a set of inputs $x$ that vary across a range.

```python
x_range = np.linspace(0, 25, num=21)
x_range
```

```python
viz_df = pd.DataFrame({'rainfall': x_range})
viz_df
```

Now let's calculate our predicted output for each of these input $x$ values.

```python
intercept = 4780
slope = 2.89
viz_df['predicted_river_depth'] = intercept + slope * viz_df.rainfall
viz_df
```

## Plot the predicted output

```python
sns.relplot(data = viz_df, x='rainfall', y='predicted_river_depth', kind='line', height=3)
plt.show()
```

## Plot the predicted and observed output

```python
observed_river_depths = {
    'rainfall': 
        [0.0, 1.25, 2.5, 3.75, 5.0, 6.25, 7.5, 8.75, 10.0, 
        11.25, 12.5, 13.75, 15.0, 16.25, 17.5, 18.75, 20.0,
        21.25, 22.5, 23.75, 25.0],
    'river_depth': 
        [4803, 4745, 4785, 4767, 4802, 4774, 4785, 4798, 4777,
        4836, 4812, 4791, 4819, 4800, 4834, 4847, 4878, 4819,
        4861, 4855, 4891]
}
observed_df = pd.DataFrame(observed_river_depths)
```

```python
sns.relplot(data = observed_df, x='rainfall', y='river_depth', kind='line', height=3)
plt.show()
```

```python
viz_df['river_depth'] = observed_df['river_depth']
viz_df
```

Let's plot the observed and predicted data on the same plot to visualize the regression's predictions as a "best fit line".

```python
fig, ax = plt.subplots(figsize=(4,3))
sns.scatterplot(data=viz_df, x='rainfall', y='river_depth', ax=ax)
sns.lineplot(data=viz_df, x='rainfall', y='predicted_river_depth', ax=ax)
plt.show()
```
