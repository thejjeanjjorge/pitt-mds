Create a function that adds two numbers.

```python
def add_numbers(a=int, b=int):
    return a+b
```

```python
add_numbers(2, 3)
```

Do the same for -, /, *

```python
def mult_numbers(a=int, b=int):
    return a*b

def sub_numbers(a=int, b=int):
    return a-b

def div_numbers(a=int, b=int):
    return a/b
```

```python
mult_numbers(2, 3)
```

```python
sub_numbers(3, 2)
```

```python
div_numbers(4, 2)
```

```python
def cal(a, b, c):
    if c == "+":
        return add_numbers(a, b)
    elif c == "-":
        return sub_numbers(a, b)
    elif c == "*":
        return mult_numbers(a, b)
    elif c == "/":
        return div_numbers(a, b)
    else:
        print("You did not select a valid option")
```

```python
cal(2, 3, "-")
```

Now doing the calculator using a While loop.

```python
def calc(a, b, c):
    if c == "+":
        return a + b
    elif c == "-":
        return a - b
    elif c == "*":
        return a * b
    elif c == "/":
        return a / b
    else:
        return None 
```

```python
while True:
    oper = input("Please enter the operator")
    a = int(input("Please the first number"))
    b = int(input("Please enter the second number"))
    result = calc(a, b, oper)

    if result is not None:
        print(f"Result {result}")
        break
    else:
        print("You did not select a valid option. Please Try again.")
```
